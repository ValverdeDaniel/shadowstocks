```python
import pandas as pd
import json
from urllib.request import urlopen
import certifi

# Function to fetch and parse JSON data
def get_jsonparsed_data(url):
    response = urlopen(url, cafile=certifi.where())
    data = response.read().decode("utf-8")
    return json.loads(data)

# API URL
url = "https://financialmodelingprep.com/stable/analyst-estimates?symbol=GM&period=annual&apikey=OknrfP7kjQPfzNSDXrhmUsf1wn9CcHZo"

# Load data
data = get_jsonparsed_data(url)

# Convert to DataFrame
df = pd.DataFrame(data)

# Convert large monetary values to billions for readability
columns_to_scale = [
    'revenueLow', 'revenueHigh', 'revenueAvg',
    'ebitdaLow', 'ebitdaHigh', 'ebitdaAvg',
    'ebitLow', 'ebitHigh', 'ebitAvg',
    'netIncomeLow', 'netIncomeHigh', 'netIncomeAvg',
    'sgaExpenseLow', 'sgaExpenseHigh', 'sgaExpenseAvg'
]

for col in columns_to_scale:
    if col in df.columns:
        df[col] = df[col] / 1e9  # Convert to billions

# Sort by date descending
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date", ascending=False)

# Show all columns including epsAvg, epsHigh, epsLow
pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", '{:,.3f}'.format)

display(df[['symbol', 'date', 'epsAvg', 'epsHigh', 'epsLow', 'numAnalystsEps']])
```
