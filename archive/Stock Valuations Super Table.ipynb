```python
df['ATH'] = df['price'].max()

df['Recent_High_120'] = df['price'].rolling(window=120, min_periods=1).max()
display(df)

first_row_dict = df.iloc[0:3].to_dict()
display(first_row_dict)
```

```python
# Iterate through each row to find recent high and all-time high
for index, row in df.iterrows():
    ticker = row["ticker"]
    current_date = row["date"]
    
    # Filter data up to current date within the last 120 days
    window_df = df[(df["ticker"] == ticker) & (df["date"] < current_date) & (df["date"] >= current_date - pd.Timedelta(days=120))]
    
    # Find the highest price in the last 120 days before the current date
    if not window_df.empty:
        recent_high_idx = window_df["price"].idxmax()
        df.at[index, "Recent_High_120"] = window_df.loc[recent_high_idx, "price"]
        df.at[index, "Recent_High_Date"] = window_df.loc[recent_high_idx, "date"]
    
    # Find all-time high for the stock up to the current date
    all_time_high_df = df[(df["ticker"] == ticker) & (df["date"] <= current_date)]
    all_time_high_idx = all_time_high_df["price"].idxmax()
    df.at[index, "All_Time_High"] = all_time_high_df.loc[all_time_high_idx, "price"]
    df.at[index, "All_Time_High_Date"] = all_time_high_df.loc[all_time_high_idx, "date"]

# Select the most recent row (today's data) to display
latest_row = df.iloc[-1:][["ticker", "price", "Recent_High_120", "Recent_High_Date", "All_Time_High", "All_Time_High_Date"]]

# Display or return latest_row
display(latest_row)
```

```python

# Function to process the data for one ticker
def process_single_ticker(ticker):
    # Step 1: Filter data for the specific ticker
    ticker_df = df[df['ticker'] == ticker].sort_values(by='date', ascending=True)

    # Step 2: Calculate All-Time High for the ticker's price
    all_time_high = ticker_df['price'].max()

    # Step 3: Calculate Recent High (last 120 days) using rolling window
    ticker_df['Recent_High_120'] = ticker_df['price'].rolling(window=120, min_periods=1).max()

    # Step 4: Calculate Percent Change from price to Recent_High_120
    ticker_df['Percent_Change_from_Recent_High'] = ((ticker_df['price'] - ticker_df['Recent_High_120']) / ticker_df['Recent_High_120']) * 100

    # Step 5: Create a new dataframe with the necessary columns for the specific ticker
    ticker_result_df = ticker_df[['ticker', 'price', 'volume', 'Recent_High_120', 'Percent_Change_from_Recent_High']].copy()
    ticker_result_df['All_Time_High'] = all_time_high

    # Step 6: Return the result
    return ticker_result_df

# Process the data for AAPL
result_df = process_single_ticker('AAPL')

# Display the result
print(result_df)
```
