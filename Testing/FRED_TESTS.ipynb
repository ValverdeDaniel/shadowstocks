```python
import pandas_datareader.data as web
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt

# Set date range
start_date = datetime(2000, 1, 1)
end_date = datetime.today()

# FRED Series IDs and labels
fred_series = {
    'CPIAUCSL': 'CPI (All Urban Consumers)',
    'CPILFESL': 'Core CPI',
    'PPIACO': 'PPI (All Commodities)',
    'PCEPILFE': 'Core PCE',
    'PCEPI': 'PCE',
    'UNRATE': 'Unemployment Rate',
    'PAYEMS': 'Nonfarm Payrolls',
    'HOUST': 'Housing Starts',
    'HSN1F': 'New One Family Houses Sold',
    'UMCSENT': 'Consumer Sentiment Index',
    'FEDFUNDS': 'Effective Fed Funds Rate',
    'T10Y2Y': '10Y - 2Y Treasury Yield',
    'M2SL': 'M2 Money Supply',
    'RETAILSMNSA': 'Retail Sales (NSA)',
    'CPROFIT': 'Corporate Profits',
    'CE16OV': 'Civilian Employment Level',
    'AHETPI': 'Average Hourly Earnings',
    'TOTALSA': 'Total Vehicle Sales',
    'WPU0911': 'PPI - Motor Fuel (Energy Proxy)'
}

# Download and merge all data into one DataFrame
data = pd.DataFrame()

for series_id, label in fred_series.items():
    try:
        series_data = web.DataReader(series_id, 'fred', start_date, end_date)
        series_data.columns = [label]
        data = pd.concat([data, series_data], axis=1)
    except Exception as e:
        print(f"Failed to load {label} ({series_id}): {e}")

# Drop empty rows
data.dropna(how='all', inplace=True)

# Preview data
data.tail()

```
